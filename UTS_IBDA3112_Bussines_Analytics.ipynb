### Menggabungkan kolom penting yang digunakan untuk menjawab soal ke dalam Data OrderDetails

```python
#  fungis untuk menghitung total_penjumlahan
def total_penjualan():
    data_order_details['Total_Penjualan'] = data_order_details['Quantity'] * data_order_details['Unit Price']

#      fungsi untuk menghitung volume
def volume():
    data_order_details['Volume'] = data_order_details['Total_Penjualan'] - (data_order_details['Discount'] * data_order_details['Total_Penjualan'])
```

Hasil di bawah menunjukkan kita sudah berhasil melakukan penambahan kolom volume dan total_penjualan

```python
total_penjualan()
volume()
data_order_details.head(1)
```

Membuat sebuah fungsi map yang digunakan untuk mengambil kolom dari dataset lain ke dalam data order_detail dengan urutan yang benar

```python
def my_map(first_data, second_data, column_1,column_2, column_3):
    first_data[column_1] = first_data[column_2].map(second_data.set_index(column_3)[column_1].to_dict())
    return first_data

data_order_details = data_order_details.set_index('Order ID').combine_first(data_orders[['Order ID', 'Order Date', 'Customer', 'Ship Via']].set_index('Order ID')).reset_index()
my_map(data_order_details,data_products,'Category','Product','Product Name')
my_map(data_order_details,data_products,'Supplier','Product','Product Name')
my_map(data_order_details,data_customers,'Customer ID','Customer','Company Name')
my_map(data_order_details,data_customers,'Country','Customer','Company Name')
```

```python
data = data_order_details
data = data.iloc[:, [3,-3,9]]

def supplier_sultan(data1, kolom1, kolom2, kolom3):
    data1[kolom1] = data1[kolom1].dt.to_period('y')
    data2 = data1.groupby([kolom1,kolom2], as_index = False, sort=False)[kolom3].sum()
    data3 = data2.groupby([kolom1], as_index = False, sort = True)[kolom3].max()
    data3[kolom2] = data3[kolom3].map(data2.set_index(kolom3)[kolom2].to_dict())
    return data3
    
```

```python
data_supp = supplier_sultan(data, 'Order Date', 'Supplier', 'Volume')
data_supp
```

Dari hasil diatas, kita dapat melihat supplier dengan total pemasokan tertinggi setiap tahunnya

```python
sns.barplot(y="Volume", hue="Supplier", x = 'Order Date',
            data=data_supp, dodge=False)
plt.title('Data Supplier Dengan Pemasok Tertinggi selama 3 Tahun', size=20)
plt.gca().set_ylim([0,100000])
plt.xlabel('Order Date (years)', fontsize=18)
plt.ylabel('Volume', fontsize=18)
plt.legend(fontsize='x-large', title_fontsize='10', title='Supplier')
```
